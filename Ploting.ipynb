```python
x=[1,2,3]
y=[4,5,5]
plt.plot(x,y)
plt.title("Info")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
x=[40,25,20,15]
la=["Mangos","Peaches","Pears","Strawberries"]
co=["#FFBF34","#FF7F50","#228b22","#fc5a8d"]
ex=[0.2,0,0,0]
plt.pie(x,labels=la,colors=co,explode=ex,shadow=True)
plt.show()
```

```python
x=[1,2,3,4]
y=[4,2,4,3]
co=["#666547","#fb2e01","#6fcb9f","#ffe28a"]
plt.barh(x,y,color=co)
plt.title("Bar Chart")
plt.xlabel("x axis",fontsize=20)
plt.ylabel("y axis",fontsize=20)
plt.show()
```

```python
x=[1,2,3,4]
y=[1,2,3,4]
siz=[200,300,390,250]
co=["#666547","#fb2e01","#6fcb9f","#ffe28a"]
plt.scatter(x,y,color=co,s=siz,edgecolor='green' ,marker="*",linewidth=0.5)
plt.title("Scatter Chart",fontsize=25)
plt.xlabel("x axis",fontsize=20)
plt.ylabel("y axis",fontsize=20)
plt.show()
```

```python
x=[1,2,3]
y=[4,5,5]
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("Info")
plt.xlabel("x axis")
plt.ylabel("y axis")



x=[40,25,20,15]
la=["Mangos","Peaches","Pears","Strawberries"]
co=["#FFBF34","#FF7F50","#228b22","#fc5a8d"]
ex=[0.2,0,0,0]
plt.subplot(2,2,2)
plt.pie(x,labels=la,colors=co,explode=ex,shadow=True)



x=[1,2,3,4]
y=[4,2,4,3]
co=["#666547","#fb2e01","#6fcb9f","#ffe28a"]
plt.subplot(2,2,3)
plt.barh(x,y,color=co)
plt.title("Bar Chart")
plt.xlabel("x axis")
plt.ylabel("y axis")


x=[1,2,3,4]
y=[1,2,3,4]
siz=[200,300,390,250]
co=["#666547","#fb2e01","#6fcb9f","#ffe28a"]
plt.subplot(2,2,4)
plt.scatter(x,y,color=co,s=siz,edgecolor='green' ,marker="*",linewidth=0.5)
plt.title("Scatter Chart")
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.subplots_adjust(left=0.1,bottom=0.1,right=0.9,top=0.9,wspace=0.4,hspace=0.4)
plt.show()
```

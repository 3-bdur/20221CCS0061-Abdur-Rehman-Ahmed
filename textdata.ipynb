```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter

text_data = "the man of"
WordCloud = WordCloud(width=800, height=400, background_color='Black').generate(text_data)

plt.figure(figsize=(10,5))
plt.bar(Counter(text_data).keys(), Counter(text_data).values())
plt.title("Histogram of word frequencies")
plt.xlabel("words")
plt.ylabel("frequency")
plt.xticks(rotation=90)
plt.show()
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

def remove_stopwords(string):
  stop_words = set(stopwords.words('english'))
  words = string.split()
  filtered_words = (word for word in words if word.lower() not in stop_words)
  return ' '.join(filtered_words)

input_string = 'this is an example sentence to remove stop words from'
result = remove_stopwords(input_string)
print("Original string:", input_string)
print("Modified string:", result)
```
